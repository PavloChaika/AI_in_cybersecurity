## Exploring and Selecting Features

In this code snippet, we perform several operations related to feature encoding and selection for a DataFrame (`df`). Here's a brief overview of the steps:

1. **Label Encoding:** We use the `LabelEncoder` from scikit-learn to numerically encode categorical columns - 'proto', 'service', and 'Attack_type'. This is done to convert textual data into a format suitable for machine learning algorithms.

2. **Correlation Analysis:** A correlation matrix is computed for the numeric columns in the DataFrame. This matrix helps identify relationships between different numerical features.

3. **Heatmap Visualization:** A heatmap is generated using `seaborn` to visually represent the correlation matrix. This is useful for identifying patterns and dependencies among variables.

4. **Correlation Thresholding:** We set a threshold (0.8 in this case) to identify highly correlated and low correlated pairs of variables. This step is crucial for feature selection, as highly correlated variables might provide redundant information.

5. **Variable Selection:** Based on the correlation analysis, a list of unique variables to keep is created. The selection process ensures that only one variable from highly correlated pairs and both variables from low correlated pairs are retained.

6. **New DataFrame Creation:** Finally, a new DataFrame (`selected_df`) is created, containing only the selected variables. But we still be using all entirety of variables, since we have ebough data to mitigate multi coliniarity"

7. **Print Selected DataFrame Size:** The number of selected variables is printed to provide insight into the dimensionality reduction achieved through the feature selection process.

This feature engineering and selection process contribute to improving the efficiency and effectiveness of machine learning models by focusing on relevant and non-redundant information.


Visualize correlation matrix

```python
label_encoder = LabelEncoder()
df['proto_encoded'] = label_encoder.fit_transform(df['proto'])
df['service_encoded'] = label_encoder.fit_transform(df['service'])
df['Attack_type_encoded'] = label_encoder.fit_transform(df['Attack_type'])

numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_columns].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
correlation_threshold = 0.8

highly_correlated_pairs = [(i, j) for i in range(correlation_matrix.shape[0]) 
                           for j in range(i + 1, correlation_matrix.shape[1]) 
                           if abs(correlation_matrix.iloc[i, j]) > correlation_threshold]

low_correlated_pairs = [(i, j) for i in range(correlation_matrix.shape[0]) 
                           for j in range(i + 1, correlation_matrix.shape[1]) 
                           if abs(correlation_matrix.iloc[i, j]) < correlation_threshold]

# Create a list of unique variables to keep
variables_to_keep = set()
for pair in highly_correlated_pairs:
    variables_to_keep.add(pair[0])  # Keep the first variable in the pair
    
for pair in low_correlated_pairs:
    variables_to_keep.add(pair[0])  # Keep the first variable in the pair
    variables_to_keep.add(pair[1])  # Keep the second variable in the pair

# Create a new DataFrame with the selected variables
selected_df = df.iloc[:, list(variables_to_keep)]

# Print the selected DataFrame
print("Selected DataFrame:")
print(len(variables_to_keep))
```

## Neural Network Training and Evaluation Function

Here, we define a function `train_and_evaluate_model` responsible for training and evaluating a neural network model. Let's break down its key components:

1. **Early Stopping:** The training process incorporates early stopping with a patience of 5 epochs and restoration of the best weights. This helps prevent overfitting by halting training when the validation loss stops improving.

2. **Model Architecture:** The neural network is a 3-layer Multilayer Perceptron (MLP) with ReLU activation and dropout after each layer. The output layer uses the softmax activation function to predict the class probabilities.

3. **Compilation:** The model is compiled with the sparse categorical crossentropy loss function, chosen for multi-class classification. The specified optimizer, batch size, and metrics (accuracy) are utilized.

4. **Training:** The model is trained for a maximum of 10,000 epochs, but early stopping may halt the training earlier. The training history is stored for later analysis.

5. **Accuracy and Loss Plots:** Plots depicting training and validation accuracy, as well as training and validation loss, are generated and saved as images in the 'plots' directory. These plots offer visual insights into the model's learning process.

6. **Evaluation:** The trained model is evaluated on the test set, and the test accuracy is printed.

7. **Return Values:** The function returns a dictionary containing relevant information such as training accuracy, validation accuracy, test accuracy, the number of epochs, and paths to the accuracy and loss plots.

This modular and organized approach facilitates experimentation with various hyperparameter configurations and provides a clear overview of the model's performance during training and evaluation.


```python
def train_and_evaluate_model(x_train, y_train, x_val, y_val, x_test, y_test,
                              optimizer='adam', batch_size=32, activation_function='relu',
                              hidden_units=128, dropout=0.3):
    
    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    
    # model is a 3-layer MLP with ReLU and dropout after each layer
    model = Sequential()
    model.add(Dense(hidden_units))
    model.add(Activation(activation_function))

    model.add(Dropout(dropout))
    model.add(Dense(hidden_units))
    model.add(Activation(activation_function))

    model.add(Dropout(dropout))
    model.add(Dense(len(np.unique(y))))
    model.add(Activation('softmax'))

    model.compile(loss='sparse_categorical_crossentropy', 
                  optimizer=optimizer,
                  metrics=['accuracy'])


    history = model.fit(x_train, y_train, epochs=10000, validation_data=(x_val, y_val), callbacks=[early_stopping])

    # Plot training & validation accuracy values
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(['Train', 'Validation'], loc='upper left')
    accuracy_plot_path = f'plots/accuracy_plot_{i + 1}.png' 
    plt.savefig(accuracy_plot_path)
    plt.close() 

    # Plot training & validation loss values
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train', 'Validation'], loc='upper left')
    loss_plot_path = f'plots/loss_plot_{i + 1}.png'
    plt.savefig(loss_plot_path)
    plt.close()

    loss, acc = model.evaluate(x_test, y_test, batch_size=batch_size)
    print("\nTest accuracy: %.1f%%" % (100.0 * acc))
    
    
    return {
        "training_accuracy": history.history['accuracy'][-1],
        "validation_accuracy": history.history['val_accuracy'][-1],
        "test_accuracy": acc,
        "num_epochs": len(history.history['accuracy']),
        "accuracy_plot_path": accuracy_plot_path,
        "loss_plot_path": loss_plot_path
        }
```

## HTML Report Generation

In this code block, an HTML report is dynamically generated based on the results of training and evaluating the neural network model with different configurations. Let's break down the key components:

1. **HTML Structure:** The report starts with the opening `<html>` and `<body>` tags.

2. **Loop Through Configurations:** A loop iterates through different configurations of optimizers, batch sizes, activation functions, hidden units, and dropout rates.

3. **Train and Evaluate Model:** The `train_and_evaluate_model` function is called for each configuration, and the results are stored in the `result` variable.

4. **HTML Division for Each Configuration:** For each configuration, a division (`<div>`) is created with a distinct style. Information about the configuration, including hyperparameters and results, is displayed.

5. **Embedded Plots:** The accuracy and loss plots are embedded using `<img>` tags, and their paths are linked to the respective result paths.

6. **Closing Tags:** The HTML report is closed with `</body></html>`.

This dynamic HTML report provides a clear and organized presentation of the model configurations and their corresponding results, including accuracy plots and loss plots. It is a valuable resource for analyzing and comparing the performance of different neural network configurations.


```python
with open('output.html', 'w') as html_file:
    html_file.write('<html><body>')

    # Loop through optimizer configurations
    for i in range(len(optimizer)):
        # Execute the model training and evaluation
        result = train_and_evaluate_model(x_train, y_train, x_val, y_val, x_test, y_test,
                                          optimizer=optimizer[i], batch_size=batch_size[i], 
                                          activation_function=activation_function[i],
                                          hidden_units=hidden_units[i], dropout=dropout[i])

        html_file.write('<div style="margin-bottom: 20px; padding: 10px; border: 1px solid #ccc; border-radius: 10px; display: flex; flex-wrap: wrap;">')
        html_file.write(f'<div style="flex: 1; margin-right: 20px;">')
        html_file.write(f'<h3 style="color: #336699;">Configuration {i + 1}</h3>')
        html_file.write('<ul style="list-style-type: none; padding: 0;">')
        html_file.write(f'<li><strong>Optimizer:</strong> {optimizer[i]}</li>')
        html_file.write(f'<li><strong>Batch Size:</strong> {batch_size[i]}</li>')
        html_file.write(f'<li><strong>Activation Function:</strong> {activation_function[i]}</li>')
        html_file.write(f'<li><strong>Hidden Units:</strong> {hidden_units[i]}</li>')
        html_file.write(f'<li><strong>Dropout:</strong> {dropout[i]}</li>')
        html_file.write(f'<li><strong>Epochs:</strong> {result["num_epochs"]}</li>')
        html_file.write('</ul>')
        html_file.write('<h4 style="color: #336699;">Results</h4>')
        html_file.write(f'<p><strong>Training Accuracy:</strong> {result["training_accuracy"]}</p>')
        html_file.write(f'<p><strong>Validation Accuracy:</strong> {result["validation_accuracy"]}</p>')
        html_file.write(f'<p><strong>Test Accuracy:</strong> {result["test_accuracy"]}</p>')
        html_file.write('</div>')

        # Embed the accuracy plot
        html_file.write(f'<div style="flex: 1; margin-top: 20px;">')
        html_file.write('<h4 style="color: #336699;">Accuracy Plot</h4>')
        html_file.write(f'<img src="{result["accuracy_plot_path"]}" alt="Accuracy Plot" style="width: 100%; border: 1px solid #ddd; border-radius: 5px;">')
        html_file.write('</div>')

        # Embed the loss plot
        html_file.write(f'<div style="flex: 1; margin-top: 20px;">')
        html_file.write('<h4 style="color: #336699;">Loss Plot</h4>')
        html_file.write(f'<img src="{result["loss_plot_path"]}" alt="Loss Plot" style="width: 100%; border: 1px solid #ddd; border-radius: 5px;">')
        html_file.write('</div>')

        html_file.write('</div>')

    # Close the HTML file
    html_file.write('</body></html>')
```
